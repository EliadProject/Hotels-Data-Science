```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

# Part 1

```python
df = pd.read_csv("hotels_data.csv")
df.head()
```

```python
# 1.a
from datetime import datetime
from dateutil.parser import parse

def get_datetime(date_str):
    return datetime.strptime(date_str, '%m/%d/%Y %H:%M')

df["DayDiff"] = DataFrame([get_datetime(val) for val in df["Checkin Date"]]) - DataFrame([get_datetime(val) for val in df["Snapshot Date"]])
df.head()
```

```python
#1.b

df["WeekDay"] = DataFrame([get_datetime(val).weekday() for val in df["Checkin Date"]])
df.head()
```

```python
#1.c

df["DiscountDiff"] = df["Original Price"] - df["Discount Price"]
df.head()
```

```python
#1.d

df["DiscountPerc"] = (df["DiscountDiff"]/df["Original Price"]) * 100
df.head()
```

```python
#Save df to csv
df.to_csv("Hotels_data_Changed.csv")
```

# Part 2

```python
# 2.a

# dropping passed columns 
filtered_df = df.filter(["Hotel Name","DayDiff", "Checkin Date", "Snapshot Date", "WeekDay", "DiscountDiff", "Discount Code"], axis=1)
filtered_df.head()
#filtered_df.to_csv("Hotels_data_for_best_discount_code.csv")
```

```python
filtered_df["MaxDis"] = filtered_df.groupby(["Hotel Name","DayDiff", "Checkin Date", "Snapshot Date", "WeekDay"])["DiscountDiff"].transform('max')
#filtered_df.sort_values(by=["Hotel Name","DayDiff", "Checkin Date", "Snapshot Date", "WeekDay"],ascending=False).reset_index()

grouped_df = filtered_df[filtered_df["DiscountDiff"]==filtered_df["MaxDis"]].drop_duplicates(["Hotel Name","DayDiff", "Checkin Date", "Snapshot Date", "WeekDay"])
grouped_df.drop(labels=["MaxDis","DiscountDiff"], axis=1,inplace=True)
grouped_df.head()
grouped_df.to_csv("Hotels_data_for_best_discount_code_grouped.csv")
# Debug stuff for find the correct data 
#grouped_df.loc[(for_learn_df["Hotel Name"] == "1 Hotel Central Park") & (for_learn_df["DayDiff"] == "3 days") & (for_learn_df["Checkin Date"] == "11/14/2015 0:00") ]
```

```python
# 2.b
pred_df = grouped_df.copy()
pred_df.head()
```

```python
pred_df_float_datetime = pred_df.copy() # copy data frame to another var so original won't change

# pre process - convert columns to categorial numebr instead of strings so fit will work.
pred_df_float_datetime["CheckinDate"] = pd.Categorical(pd.factorize(pred_df_float_datetime["Checkin Date"])[0]) 
pred_df_float_datetime["SnapshotDate"] = pd.Categorical(pd.factorize(pred_df_float_datetime["Snapshot Date"])[0])
pred_df_float_datetime["HotelName"] = pd.Categorical(pd.factorize(pred_df_float_datetime["Hotel Name"])[0])



pred_df_float_datetime["DayDiff"] = [t.total_seconds() for t in pred_df_float_datetime["DayDiff"]] # convert timedelta to float

# remove target column "Discount Code" from data frame used for learning.
pred_df_float_datetime.drop(labels=["Discount Code","Checkin Date","Snapshot Date", "Hotel Name"],axis=1, inplace=True)
pred_df_float_datetime.head(5)
```

# Naive Bayes - All train All test

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(pred_df_float_datetime, pred_df['Discount Code']).predict(pred_df_float_datetime)
y_pred, pred_df['Discount Code']
print("Number of mislabeled points out of a total %d points : %d    -> error : %f%%" 
      % (pred_df_float_datetime.shape[0],(pred_df['Discount Code'] != y_pred).sum(),
        (pred_df['Discount Code'] != y_pred).sum()/pred_df_float_datetime.shape[0]*100))
```

# Cross Validation

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(pred_df_float_datetime, pred_df['Discount Code'], test_size=0.3,random_state=109) # 70% training and 30% test

X_train, X_test, y_train, y_test

```

# Naive Bayes - 0.3/0.7

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
gnb = GaussianNB()

#Train the model using the training sets
gnb.fit(X_train, y_train)

#Predict the response for test dataset
y_pred2 = gnb.predict(X_test)
y_pred2

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred2))
```

# ROC

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

#def myfunc(a):
#    l = [0,0,0,0]
#    l[a-1] = 1
#    return l

#y_test_probs = map(myfunc, y_test.tolist())
#print(y_test_probs)

#y_pred3_probs = map(myfunc, y_pred3.tolist())
#print(y_pred3_probs)

y = label_binarize(pred_df['Discount Code'], classes=pred_df['Discount Code'].unique().tolist())
#print(y)
n_classes = y.shape[1]

#print(list(y_test_probs))

for i in range(n_classes):
    #fpr[i], tpr[i], _ = roc_curve(np.array(list(y_test_probs))[:,i], np.array(list(y_pred3_probs))[:,i], pos_label=4)
    fpr[i], tpr[i], _ = roc_curve(y_test[:], y_pred3[:], pos_label=4)
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_pred3.ravel(), pos_label=4)
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

fpr
```

```python
import matplotlib.pyplot as plt
plt.figure()
lw = 2
plt.plot(fpr[0], tpr[0], color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[0])
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
