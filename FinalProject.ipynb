```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

# Part 1

```python
df = pd.read_csv("hotels_data.csv")
df.head()
```

```python
# 1.a
from datetime import datetime
from dateutil.parser import parse

def get_datetime(date_str):
    return datetime.strptime(date_str, '%m/%d/%Y %H:%M')

df["DayDiff"] = DataFrame([get_datetime(val) for val in df["Checkin Date"]]) - DataFrame([get_datetime(val) for val in df["Snapshot Date"]])
df.head()
```

```python
#1.b

df["WeekDay"] = DataFrame([get_datetime(val).weekday() for val in df["Checkin Date"]])
df.head()
```

```python
#1.c

df["DiscountDiff"] = df["Original Price"] - df["Discount Price"]
df.head()
```

```python
#1.d

df["DiscountPerc"] = (df["DiscountDiff"]/df["Original Price"]) * 100
df.head()
```

```python
#Save df to csv
df.to_csv("Hotels_data_Changed.csv")
```

# Part 2

```python
# 2.a

# dropping passed columns 
pred_df = df.filter(["Hotel Name","DayDiff", "Checkin Date", "Snapshot Date", "WeekDay", "DiscountDiff", "Discount Code"], axis=1)
pred_df.head()
#pred_df.to_csv("Hotels_data_for_best_discount_code.csv")
```

```python
pred_df["MaxDis"] = pred_df.groupby(["Hotel Name","DayDiff", "Checkin Date", "Snapshot Date", "WeekDay"])["DiscountDiff"].transform('max')
#pred_df.sort_values(by=["Hotel Name","DayDiff", "Checkin Date", "Snapshot Date", "WeekDay"],ascending=False).reset_index()

for_learn_df = pred_df[pred_df["DiscountDiff"]==pred_df["MaxDis"]].drop_duplicates(["Hotel Name","DayDiff", "Checkin Date", "Snapshot Date", "WeekDay"])
for_learn_df.drop(labels=["MaxDis","DiscountDiff"], axis=1,inplace=True)
for_learn_df

# Debug stuff for find the correct data 
#for_learn_df.loc[(for_learn_df["Hotel Name"] == "1 Hotel Central Park") & (for_learn_df["DayDiff"] == "3 days") & (for_learn_df["Checkin Date"] == "11/14/2015 0:00") ]
```

```python
# 2.b
pred_df = for_learn_df
pred_df.head()
hotels_count = len(pred_df["Hotel Name"].unique())
day_diff_count = len(pred_df["DayDiff"].unique())
checkin_date_count = len(pred_df["Checkin Date"].unique())
snapshot_date_count = len(pred_df["Snapshot Date"].unique())
week_day_count = len(pred_df["WeekDay"].unique())

hotels_count, day_diff_count, checkin_date_count, checkin_date_count, snapshot_date_count, week_day_count
```

```python
pred_df.head()

```

```python
pred_df_float_datetime = pred_df.copy() # copy data frame to another var so original won't change

# pre process - convert columns to categorial numebr instead of strings so fit will work.
pred_df_float_datetime["Checkin Date"] = pd.Categorical(pd.factorize(pred_df_float_datetime["Checkin Date"])[0]) 
pred_df_float_datetime["Snapshot Date"] = pd.Categorical(pd.factorize(pred_df_float_datetime["Snapshot Date"])[0])
pred_df_float_datetime["Hotel Name"] = pd.Categorical(pd.factorize(pred_df_float_datetime["Hotel Name"])[0])

pred_df_float_datetime["DayDiff"] = [t.total_seconds() for t in pred_df_float_datetime["DayDiff"]] # convert timedelta to float

# remove target column "Discount Code" from data frame used for learning.
pred_df_float_datetime.drop(labels=["Discount Code"],axis=1, inplace=True)
pred_df_float_datetime.head(5)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(pred_df_float_datetime, pred_df['Discount Code']).predict(pred_df_float_datetime)
y_pred
print("Number of mislabeled points out of a total %d points : %d    -> error : %f%%" 
      % (pred_df_float_datetime.shape[0],(pred_df['Discount Code'] != y_pred).sum(),
        (pred_df['Discount Code'] != y_pred).sum()/pred_df_float_datetime.shape[0]*100))
```
